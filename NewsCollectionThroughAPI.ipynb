```python
import pprint
import requests     # 2.19.1
secret = '548e22f631e'
url = 'https://newsapi.org/v2/everything?'
parameters = {
    'q': 'Ministry of truth',
    'pageSize': 100,  # max 100
    'apiKey': secret
}
# Make the request
response = requests.get(url, params=parameters)

# Convert the response to JSON format and pretty print it
response_json = response.json()
pprint.pprint(response_json)
```

```python
import pandas as pd
author,content,description,publishedAt,source,title,url = [],[],[],[],[],[],[]
for article in response_json['articles']:
    author.append(article[u'author'])
    content.append(article[u'content'])
    description.append(article[u'description'])
    publishedAt.append(article[u'publishedAt'])
    source.append(article[u'source'][u'name'])
    title.append(article[u'title'])
    url.append(article[u'url'])
df = pd.DataFrame([author,content,description,publishedAt,source,title,url]).T
df.columns =['author','content',"description","publishedAt","source","title","url"]
df.head(5)
```
